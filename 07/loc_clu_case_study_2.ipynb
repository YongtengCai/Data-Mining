```python
# 读取评论数据集

review_data = pd.read_csv("./data/review_data.csv")
review_data.head()
```

```python
# 读取餐馆数据集
biz_data = pd.read_csv("./biz_data.csv")
```

### 1. 数据处理

```python
# 获取用户访问数据
data1 = review_data[['user_id', 'business_id', 'date']]
## 转换日期格式
date = pd.to_datetime(data1["date"], format = '%Y-%m-%d')
data1['date'] = date


# 获取餐馆经纬度信息
data2 = biz_data[['business_id', 'latitude', 'longitude']]
```

```python
# 将用户访问数据与餐馆地理位置数据合并
data = pd.merge(data1, data2, on = 'business_id')
data.head()
```

```python
data = data.sort_values(by = ['user_id', 'date'], ascending = True)

# 将合并后的数据存入文件user_trajectories中
data.to_csv("./data/user_trajectories.csv", index = False)
```

```python
data = pd.read_csv("./data/user_trajectories.csv")
print("Total rows: ", data.shape[0])
data.head()
```

```python
## 检查缺失情况
missing_values_count = data.isnull().sum()
print(missing_values_count)
```

```python
# 统计每个用户访问餐馆的次数情况
user_count = data["user_id"].value_counts()
print(user_count)

user_list=list(user_count.index)
```

### 2. 用户访问轨迹可视化

```python
import folium
import os
```

```python
# 按用户id获取用户访问餐馆轨迹，然后可视化

# Create map and display it
# 多伦多经纬度
latitude = 43.66
longitude = -79.38

trt_map = folium.Map(location = [latitude, longitude], zoom_start = 12)

color = ['royalblue', 'red', 'green', 'yellow', 'purple']
i = -1

# 对第500到505位用户进行可视化
for user_id in user_list[500:505]:
    i = i + 1
    user_data = data[data['user_id'] == user_id]
    trajectory = []
    
    for name, row in user_data.iterrows():
        trajectory.append([row['latitude'],row['longitude']]) 
        folium.CircleMarker( 
            [row['latitude'],row['longitude']],
            radius = 4,  # define how big you want the circle markers to be
            color = 'grey',
            fill = False).add_to(trt_map)      
    folium.PolyLine(trajectory, weight = 3, opacity = 0.6, color = color[i]).add_to(trt_map)

trt_map
```
